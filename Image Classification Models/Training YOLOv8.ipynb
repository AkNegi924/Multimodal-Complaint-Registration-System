# Training YOLOv8

```python
from ultralytics import YOLO
import os
import cv2
import imghdr
```

```python
data_dir="data1"
image_exts=['jpeg','jpg','bmp','png']
```

```python
for typ in ['train', 'val']:
  for image_class in os.listdir(os.path.join(data_dir,typ)):
    for image in os.listdir(os.path.join(data_dir,typ,image_class)):
      image_path=os.path.join(data_dir, typ,image_class, image)
      try:
        img=cv2.imread(image_path)
        tip=imghdr.what(image_path)
        if tip not in image_exts:
          print('Image not in ext list {}'.format(image_path))
          os.remove(image_path)
      except Exception as e:
        print('Issue with image {}'.format(image_path))
        os.remove(image_path)
```

```python
model = YOLO('yolov8n-cls.pt')  # load a pretrained model (recommended for training)
results = model.train(data='data1', epochs=10, imgsz=64)
```

```python
model = YOLO(r'runs\classify\train5\weights\last.pt')  # load a custom model
results = model(r'test\Water2.jpeg')
print(results)
```

```python
names_dict=results[0].names
# probab=results[0].probs.tolist()
for result in results:
  boxes = result.boxes  # Boxes object for bounding box outputs
  print(type(boxes))
  masks = result.masks  # Masks object for segmentation masks outputs
  print(type(masks))
  keypoints = result.keypoints  # Keypoints object for pose outputs
  print(type(keypoints))
  probs = result.probs  # Probs object for classification outputs
  print(type(probs))
  result.show()  # display to screen
  result.save(filename='result.jpg')  # save to disk
```
